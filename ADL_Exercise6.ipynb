```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE, BorderlineSMOTE
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
# Load data
data = pd.read_csv('Imbalanced_data.csv')

# show
print(data.head())
```

```python
# check data
X = data.iloc[:, :-1].values  # coordinate-feature
y = data.iloc[:, -1].values   # label

# Class Distribution
print(f"Class Distribution: {Counter(y)}")
```

```python
# plot func
def plot_data(X, y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.6)
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.6)
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python
plot_data(X, y, "Original Dataset")
```

```python
# data augmentation with SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
print(f"SMOTE class distribution: {Counter(y_smote)}")
plot_data(X_smote, y_smote, "Dataset After SMOTE")
```

```python
# 3. data augmentation with Borderline-SMOTE
borderline_smote = BorderlineSMOTE(random_state=42, kind='borderline-1')
X_borderline, y_borderline = borderline_smote.fit_resample(X, y)

print(f"Borderline-SMOTE distribution : {Counter(y_borderline)}")
plot_data(X_borderline, y_borderline, "Dataset After Borderline-SMOTE")
```

```python
# Compare
print("Orijinal dataset:", Counter(y))
print("SMOTE dataset:", Counter(y_smote))
print("Borderline-SMOTE dataset:", Counter(y_borderline))
```
